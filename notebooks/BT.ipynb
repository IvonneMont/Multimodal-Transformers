```python
data =[json.loads(l) for l in open("data_sets/proc/moviescope/train.jsonl")]
```

```python
data_aug=[]
```

```python
i=0
while i<len(data):
    print("data", i)
    j=0
    sentences=data[i]['synopsis'].split(". ")
    s_aug=""
    for s in sentences:
        print(j)
        j+=1
        if len(s)==0:
            continue
        result = trans.translate(s, src='en', tmp = 'fr')
        s_aug+=result.result_text +". "
    data_aug.append(s_aug)
    i+=1
```
