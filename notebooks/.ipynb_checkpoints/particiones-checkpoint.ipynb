```python
path_img=glob.glob("data_sets/proc/moviescope/sinteticas2/*")
```

```python
for p in path_img:
    g=os.path.split(p)[-1].replace("_"," ").replace(".jpg","").split(" ")[-1]
    if g =="Sci-fi":
        print(p)
```

```python
for i in range(len(path_img)):
    path_img[i]=os.path.split(path_img[i])[-1]
    
```

```python
path_img
```

```python
ids=[]
for p in path_img:
    ids.append(os.path.split(p)[-1])
```

```python
for i in range(len(ids)):
    ids[i]=ids[i].replace("_"," ").split(" ")[0]
```

```python
ids=list(set(ids))
```

```python
len(ids)
```

```python
data=[json.loads(l) for l in open("data_sets/proc/moviescope/train.jsonl")]
```

```python
data_aumentados=[]
for d in data:
    data_aumentados.append(d)
    if str(d["id"]) in ids:
        sentences=tokenizer.tokenize(d['synopsis'])
        sz=0
        ngenres=[]
        for g in d["label"]:
            if str(d["id"])+"_"+g+".jpg" in path_img:
                ngenres.append(g)
        if len(ngenres)>0:
            sz=int(len(sentences)/len(ngenres))
        i=0
        for g in ngenres:
            text=" ".join(sentences[sz*i:sz*(i+1)])
            i+=1
            aug={"id":str(d["id"])+"_"+g,"label":[g],'title':d['title'],'plot_keywords':d['plot_keywords'],'synopsis':text}
            data_aumentados.append(aug)
```

```python
len(data_aumentados)
```

```python
with jsonlines.open("data_sets/proc/moviescope/train_aug.jsonl",'w') as writer:
    writer.write_all(data_aumentados)
```
