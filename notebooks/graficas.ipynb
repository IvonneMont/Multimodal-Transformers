```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_train(file):
    m=[]
    for i in range(1,6):
        p=file+str(i)+"/metricas.jsonl"
        with open(p, "r") as read_file:
            metricas = json.load(read_file)
            m.append(metricas)
        
    mf1_train=[]
    min_l=100
    for d in m:
        a=np.array(d["mi_f1_train"])
        if len(a)<min_l:
            min_l=len(a)
        mf1_train.append(a)
    av_f1=np.zeros(min_l)
    for a in mf1_train:
        av_f1+=a[:min_l]
    av_mf1_train=av_f1/5
    
    mf1_val=[]
    min_l=100
    for d in m:
        a=np.array(d["mi_f1_val"])
        if len(a)<min_l:
            min_l=len(a)
        mf1_val.append(a)
    av_f1=np.zeros(min_l)
    for a in mf1_val:
        av_f1+=a[:min_l]
    av_mf1_val=av_f1/5

    # evenly sampled time at 200ms intervals
    t = range(len(av_mf1_train))
    mx=np.ones(len(av_mf1_train))*np.max(av_mf1_val)

    # red dashes, blue squares and green triangles
    plt.axvline(np.argmax(av_mf1_val),color='k', linestyle='--')
    plt.plot(t, av_mf1_train, 'r', t, av_mf1_val, 'b',t,mx,'b--')
    #plt.yscale('log')
    plt.show()
```
